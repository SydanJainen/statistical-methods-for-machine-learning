```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '../data/raw/dataset.csv'
```

```python
df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
df.describe()
```

```python
df = df.dropna()
plt.figure(figsize=(10, 6))
sns.histplot(df['popularity'], bins=30, kde=True)
plt.title('Distribuzione della popolarita')
plt.show()
```

```python
selected_columns = ['popularity', 'duration_ms']
df_selected = df[selected_columns]

#df_filtered = df[df['popularity'] > 50]

plt.figure(figsize=(10, 6))
sns.scatterplot(x='popularity', y='duration_ms', data=df)
plt.title('Scatter Plot')
plt.show()
```

```python
selected_columns = ['popularity', 'danceability']
df_selected = df[selected_columns]

#df_filtered = df[df['popularity'] > 50]

plt.figure(figsize=(10, 6))
sns.scatterplot(x='popularity', y='danceability', data=df)
plt.title('Scatter Plot')
plt.show()
```

```python

correlation_matrix = df.corr()
popularity_correlation = correlation_matrix['popularity'].sort_values(ascending=False)

print("Correlazione con la popularity:")
print(popularity_correlation)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='loudness', y='popularity', data=df)
plt.title('Scatter Plot tra Popularity e loudness')
plt.show()
```
